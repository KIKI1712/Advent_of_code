```python
def numbers(data):
    only_numbers = []
    for i in values:
        number = re.findall(r'\d', i)

        if len(number) == 1:
            only_numbers.append(str(number[0])+str(number[0]))
        else:
            first_nmb = number[0]
            last_nmb = number[-1]
            only_numbers.append(str(first_nmb) + str(last_nmb))
    
    only_numbers2 = []
    for n in only_numbers:
        only_numbers2.append(int(n))
    
    return sum(only_numbers2)

    
```
