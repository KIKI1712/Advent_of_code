```python
def calculate_final_position(directions):
    x, y = 0, 0  
    global polygon2
    polygon2 = []
    for direction, distance in directions:
        if direction == "R":
            x += int(distance) 
        elif direction == "L":
            x -= int(distance) 
        elif direction == "U":
            y += int(distance)
        elif direction == "D":
            y -= int(distance) 

        polygon2.append([x,y])

    #return(polygon2)

calculate_final_position(directions)
```

```python
def polygonArea(vertices):

  numberOfVertices = len(vertices)
  sum1 = 0
  sum2 = 0
  
  for i in range(0,numberOfVertices-1):
    sum1 = sum1 + (vertices[i][0] *  vertices[i+1][1])
    sum2 = sum2 + (vertices[i][1] *  vertices[i+1][0])
  
  #Add xn.y1
  sum1 = sum1 + vertices[numberOfVertices-1][0]*vertices[0][1]   
  #Add x1.yn
  sum2 = sum2 + vertices[0][0]*vertices[numberOfVertices-1][1]   
  
  area = abs(sum1 - sum2) / 2
  area = area + perimeter/2 +1
  return area
```

```python
area = polygonArea(polygon2)
print("Polygon Vertices:")
print(polygon2)
print("")
print("Area = " + str(area) + "cm2")
```
